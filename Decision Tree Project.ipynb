```python
import math
import numpy as np
def entropy(YData) :
    #print("rsthey",type(YData))
    #print(YData)
    classes = set(YData)
    num_of_data_pt = list()
    for i in classes:
        count =0
        for j in YData:
            if i==j:
                count=count+1
        num_of_data_pt.append(count)
    #print(num_of_data_pt)
    entropy = 0
    if len(classes) == 1:
        return entropy
    #print(len(classes))
    for i in range(len(classes)):
        p = num_of_data_pt[i]/len(YData)
        base = len(classes)
        #print(base,p)
        entropy += -1*p*math.log(p,base)
    return entropy
Y = np.array([1,1,2,2,1,3,4,5,6,7])
n = entropy(Y)
n
```

```python
def gain(X,Y,sel_f):
    #sel_f is the column number of feature about which calculation will be done
    sel_col = list()
    for i in range(len(X)): #extracting the column
        sel_col.append(X[i][sel_f])
    entropy_node = entropy(Y)
    entropy_children = 0
    #sel_col = tuple(sel_col)
    #print(sel_col)
    children = set(sel_col) 
    for c in children:
        Ynew = list()
        for i in range(len(Y)):
            if sel_col[i]==c:
                Ynew.append(Y[i]) 
        Ynew = tuple(Ynew)
        entropy_children+=entropy(Ynew)
        delta_g = entropy_node - entropy_children
    return delta_g, children

def typesPrint(Y):
    classes = set(Y)
    for i in classes:
        count = 0
        for j in Y:
            if j == i:
                count = count +1
        print("count of ", i, " = ",count)       
```

```python
def buildTree(level,X,Y,features): 
    #print("they",type(Y))
    #print(X)
    #print(Y)
    #print(features)
    node = DecTreeNode(level,X,Y,features)
    K = set(node.Y)
    if len(K)==1:
        return 
    if len(node.features)==0:
        return 
    maxgain = -10000
    for f in range(len(node.features)):
        temp = gain(X,Y,f) 
        if temp[0] > maxgain:
            maxgain = temp[0]
            selected_feature = f
            splits = temp[1]
    print("Splitting about feature ", node.features[selected_feature]," with increased gain = ", maxgain)
    
    for s in splits:
        X_send = list()
        Y_send = list()
        feature_send = list()
        for i in range(len(node.features)):
            if i!=selected_feature:       
                feature_send.append(node.features[i])
        #building X_send sending all columns except selcted feature and entries belonging to one class  
        for i in range(len(X)):
            X_this_class = list()
            if X[i][selected_feature] == s:
                 for j in range(len(X[i])):
                     if j!=selected_feature:
                           X_this_class.append(X[i][j])
                 Y_send.append(Y[i])
                 X_send.append(X_this_class)
        #feature_send = tuple(feature_send)
        #X_send = tuple(X_send)
        #Y_send = tuple(Y_send)
        child_node = buildTree(level+1,X_send,Y_send,feature_send)
        node.children.append(child_node)
    return node
```
